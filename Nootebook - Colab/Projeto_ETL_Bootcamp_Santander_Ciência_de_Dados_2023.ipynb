###Extract

```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import json
```

```python
dados = pd.read_csv('dados.csv')
dados
```

###Transform

```python
x_train, x_test, y_train, y_test = train_test_split(
    dados[['resistencia', 'capacitancia', 'indutancia']],
    dados['impedancia'],
    random_state=42,
    test_size=0.2
)
```

```python
modelo_arvore = DecisionTreeRegressor()
```

```python
modelo_rede = tf.keras.Sequential([
    tf.keras.layers.Dense(79, activation='relu', input_shape=(3,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(29, activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
modelo_rede.compile(optimizer='adam', loss='mse')
```

```python
modelo_arvore.fit(x_train, y_train)
```

```python
modelo_rede.fit(x_train, y_train, epochs=1000, batch_size=79, verbose=1)
```

```python
previsoes_arvore = modelo_arvore.predict(x_test)
previsoes_rede = modelo_rede.predict(x_test)
```

```python
MAE_arvore = mean_absolute_error(y_test, previsoes_arvore)
MSE_arvore = mean_squared_error(y_test, previsoes_arvore)
R2_arvore = r2_score(y_test, previsoes_arvore)

MAE_rede = mean_absolute_error(y_test, previsoes_rede)
MSE_rede = mean_squared_error(y_test, previsoes_rede)
R2_rede = r2_score(y_test, previsoes_rede)
```

```python
print('Erro médio Absoluto da Árvore de Decisão é: ',MAE_arvore)
print('Erro médio quadrático da Árvore de Decisão é: ',MSE_arvore)
print('Coeficiente de Determinação da Árvore de Decisão é: ',R2_arvore)
print('________________________________________________________')

print(f'Erro médio Absoluto da Rede Neural é : {MAE_rede}.')
print(f'Erro médio Quadrático da Rede Neural é: {MSE_rede}.')
print(f'Coeficiente de Determinação da Rede Neural é: {R2_rede}.')
```

```python
novos_dados = pd.DataFrame({'resistencia':[725.4], 'capacitancia':[62.9], 'indutancia':[47.8]})
novas_previsoes_arvore = modelo_arvore.predict(novos_dados)
novas_previsoes_rede = modelo_rede.predict(novos_dados)

print(novas_previsoes_arvore)
print(novas_previsoes_rede)
```

```python
dados = {
    'resultados_previsoes_arvore':float(novas_previsoes_arvore),
    'resultados_previsoes_rede':float(novas_previsoes_rede),
}

json_str = json.dumps(dados)

with open('dados.json', 'w') as f:
    f.write(json_str)
```
